```python
def findoutliers(df,column):
    z_score = (df[column] - df[column].mean()) / df[column].std()

    outliers = df[(z_score > 3) | (z_score < -3)]

    return len(outliers)

def findoutIQR(df,column):

    Q1= df[column].quantile(0.25)
    Q3= df[column].quantile(0.75)
    IQR = df[column].quantile(0.75) - df[column].quantile(0.25)

    outliers = df[(df[column] > Q3 + (1.5*IQR)) | (df[column] < Q3 - (1.5*IQR))]

    return len(outliers)
    
```

```python
def cap(df,column):

    Q1= df[column].quantile(0.25)
    Q3= df[column].quantile(0.75)
    IQR = Q3 - Q1

    upper_ = Q3 + (1.5*IQR) 
    lower_ = Q3 - (1.5*IQR)
    
    df[column] = df[column].clip(lower = lower_,upper = upper_)

    return df
```

```python
df = cap(df,'total_bedrooms')
df= cap(df,'population')
df=cap(df,'households')
df=cap(df,'median_income')
df=cap(df,'rooms_imputed')
```
